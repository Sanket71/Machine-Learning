# **3. OneHotEncoding using SKlearn**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=2)
```

The random_state parameter in train_test_split is used for reproducibility. When you set random_state to an integer, it ensures that the data split will be the same every time you run your code. If you don't set random_state, the split will be different each time, which can make it hard to compare results between runs.

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

drop='first': This parameter is used to prevent multicollinearity, which can be an issue in some machine learning models. When set to 'first', it drops the first category for each feature being encoded. For example, if your 'fuel' column has 'Petrol', 'Diesel', and 'CNG', and 'Petrol' is the first category encountered, then OneHotEncoder will create columns for 'Diesel' and 'CNG' but not 'Petrol'. If both 'Diesel' and 'CNG' are 0, it implicitly means the fuel type is 'Petrol'. This is also known as K-1 encoding.

sparse_output=False: By default, OneHotEncoder returns a sparse matrix, which is efficient for high-dimensional data with many zeros. However, by setting sparse_output=False, you instruct the encoder to return a dense NumPy array instead. This can be more convenient for certain operations or if your dataset is not extremely sparse.

dtype=np.int32: This parameter specifies the data type for the output array. By setting it to np.int32, the encoded values (0s and 1s) will be stored as 32-bit integers. This can help manage memory usage and ensures type consistency if you're integrating with other NumPy operations or machine learning libraries.
